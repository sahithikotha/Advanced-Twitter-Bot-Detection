```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Embedding
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('C:\Users\sk304939\OneDrive - California State University, Northridge\Documents\Thesis\My_Project\Twitter Bot - Final code\Previous EDA code\bot_detection_data.csv')  # Load the correct file path

# Extracting the Tweet text and Bot Label
tweets = data['Tweet'].values
bot_labels = data['Bot Label'].values

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    tweets, bot_labels, test_size=0.2, random_state=42, stratify=bot_labels
)

# Tokenization and padding
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(X_train)

# Convert the text to sequences
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Padding the sequences to ensure uniform input length
max_sequence_length = 100  # Maximum length of sequences after padding
X_train_pad = pad_sequences(X_train_seq, maxlen=max_sequence_length)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_sequence_length)

# Print the shapes to debug
print(f"Training data shape: {X_train_pad.shape}")
print(f"Test data shape: {X_test_pad.shape}")

# Build LSTM model
def create_lstm_model(input_shape):
    model = Sequential()

    # Embedding layer (input length defined by max_sequence_length)
    model.add(Embedding(input_dim=5000, output_dim=100, input_length=input_shape[1]))

    # First LSTM layer with return_sequences=True for the next LSTM layer
    model.add(LSTM(64, return_sequences=True))
    model.add(Dropout(0.3))

    # Second LSTM layer
    model.add(LSTM(32, return_sequences=False))  # Only last LSTM layer needs return_sequences=False
    model.add(Dropout(0.3))

    # Dense layer
    model.add(Dense(16, activation='relu'))
    model.add(Dropout(0.2))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model

# Create model
lstm_model = create_lstm_model((1, X_train_pad.shape[1]))  # Input shape from padded sequences
lstm_model.summary()  # This should now print the correct output shapes

# Train LSTM model
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=10, restore_best_weights=True
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=5, min_lr=0.0001
)

history = lstm_model.fit(
    X_train_pad, y_train,
    batch_size=32,
    epochs=100,
    validation_data=(X_test_pad, y_test),
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# Evaluate LSTM model
lstm_train_pred_proba = lstm_model.predict(X_train_pad)
lstm_test_pred_proba = lstm_model.predict(X_test_pad)

lstm_train_pred = (lstm_train_pred_proba > 0.5).astype(int).flatten()
lstm_test_pred = (lstm_test_pred_proba > 0.5).astype(int).flatten()

# Performance metrics
print("LSTM Model Performance:")
print("="*50)
print("Training Results:")
print(f"Accuracy: {accuracy_score(y_train, lstm_train_pred):.4f}")
print(f"Precision: {precision_score(y_train, lstm_train_pred):.4f}")
print(f"Recall: {recall_score(y_train, lstm_train_pred):.4f}")
print(f"F1-Score: {f1_score(y_train, lstm_train_pred):.4f}")

print("\nTest Results:")
print(f"Accuracy: {accuracy_score(y_test, lstm_test_pred):.4f}")
print(f"Precision: {precision_score(y_test, lstm_test_pred):.4f}")
print(f"Recall: {recall_score(y_test, lstm_test_pred):.4f}")
print(f"F1-Score: {f1_score(y_test, lstm_test_pred):.4f}")

print(f"Classification Report:\n{classification_report(y_test, lstm_test_pred)}")

# Plot training history
plt.figure(figsize=(15, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('LSTM Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('LSTM Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Confusion Matrix for LSTM
lstm_cm = confusion_matrix(y_test, lstm_test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(lstm_cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Human', 'Bot'],
            yticklabels=['Human', 'Bot'])
plt.title('LSTM Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout, Embedding
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('C:\Users\sk304939\OneDrive - California State University, Northridge\Documents\Thesis\My_Project\Twitter Bot - Final code\Previous EDA code\bot_detection_data.csv')  # Ensure the correct path is used

# Extracting the Tweet text and Bot Label
tweets = data['Tweet'].values
bot_labels = data['Bot Label'].values

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    tweets, bot_labels, test_size=0.2, random_state=42, stratify=bot_labels
)

# Tokenization and padding
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(X_train)

# Convert the text to sequences
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Padding the sequences to ensure uniform input length
max_sequence_length = 100  # Maximum length of sequences after padding
X_train_pad = pad_sequences(X_train_seq, maxlen=max_sequence_length)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_sequence_length)

# Print the shapes to debug
print(f"Training data shape: {X_train_pad.shape}")
print(f"Test data shape: {X_test_pad.shape}")

# Build GRU model
def create_gru_model(input_shape):
    model = Sequential()

    # Embedding layer (input length defined by max_sequence_length)
    model.add(Embedding(input_dim=5000, output_dim=100, input_length=input_shape[1]))

    # First GRU layer with return_sequences=True for the next GRU layer
    model.add(GRU(64, return_sequences=True))
    model.add(Dropout(0.3))

    # Second GRU layer
    model.add(GRU(32, return_sequences=False))  # Only last GRU layer needs return_sequences=False
    model.add(Dropout(0.3))

    # Dense layer
    model.add(Dense(16, activation='relu'))
    model.add(Dropout(0.2))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model

# Create GRU model
gru_model = create_gru_model((1, X_train_pad.shape[1]))  # Input shape from padded sequences
gru_model.summary()  # This should now print the correct output shapes

# Train GRU model
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=10, restore_best_weights=True
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=5, min_lr=0.0001
)

history = gru_model.fit(
    X_train_pad, y_train,
    batch_size=32,
    epochs=100,
    validation_data=(X_test_pad, y_test),
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# Evaluate GRU model
gru_train_pred_proba = gru_model.predict(X_train_pad)
gru_test_pred_proba = gru_model.predict(X_test_pad)

gru_train_pred = (gru_train_pred_proba > 0.5).astype(int).flatten()
gru_test_pred = (gru_test_pred_proba > 0.5).astype(int).flatten()

# Performance metrics
print("GRU Model Performance:")
print("="*50)
print("Training Results:")
print(f"Accuracy: {accuracy_score(y_train, gru_train_pred):.4f}")
print(f"Precision: {precision_score(y_train, gru_train_pred):.4f}")
print(f"Recall: {recall_score(y_train, gru_train_pred):.4f}")
print(f"F1-Score: {f1_score(y_train, gru_train_pred):.4f}")

print("\nTest Results:")
print(f"Accuracy: {accuracy_score(y_test, gru_test_pred):.4f}")
print(f"Precision: {precision_score(y_test, gru_test_pred):.4f}")
print(f"Recall: {recall_score(y_test, gru_test_pred):.4f}")
print(f"F1-Score: {f1_score(y_test, gru_test_pred):.4f}")

print(f"Classification Report:\n{classification_report(y_test, gru_test_pred)}")

# Plot training history
plt.figure(figsize=(15, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('GRU Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('GRU Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Confusion Matrix for GRU
gru_cm = confusion_matrix(y_test, gru_test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(gru_cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Human', 'Bot'],
            yticklabels=['Human', 'Bot'])
plt.title('GRU Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
